```python
# 0. Êï∞ÊçÆÊ∏ÖÊ¥ó (Èò≤ÁàÜÁõæÁâå) - ÊîæÂú®ÊúÄÂâçÈù¢Ë∑ë‰∏ÄÊ¨° [PARALLEL VERSION]

def process_single_file(file_info):
    """Process a single file (used for parallel processing)"""
    file_path, valid_extensions = file_info
    file = os.path.basename(file_path)
    
    result = {'converted': 0, 'deleted': 0, 'warning': None}
    
    # 1. Âà†Èô§ÈùûÂõæÁâáÊñá‰ª∂
    if not file.lower().endswith(valid_extensions):
        if not file.startswith('.'):
            result['warning'] = f"‚ö†Ô∏è Non-image file found: {file}"
        return result
    
    # 2. Ê£ÄÊü•ÂõæÁâáËÉΩÂê¶ÊâìÂºÄ & Âº∫Âà∂ËΩ¨‰∏∫ RGB
    try:
        with Image.open(file_path) as img:
            if img.mode != 'RGB':
                img = img.convert('RGB')
                img.save(file_path)
                result['converted'] = 1
    except Exception as e:
        result['warning'] = f"‚ùå Corrupted file found and removed: {file_path} ({e})"
        try:
            os.remove(file_path)
            result['deleted'] = 1
        except:
            pass
    
    return result

def clean_and_verify_data(root_dir, max_workers=None):
    """Parallel version of data cleaning using ThreadPoolExecutor"""
    print(f"üßπ Scanning and cleaning dataset at: {root_dir} ...")
    
    valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp')
    file_list = []

    # 1. Collect all file paths
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            file_path = os.path.join(subdir, file)
            file_list.append((file_path, valid_extensions))

    print(f"üìÅ Found {len(file_list)} files to process")
    
    if max_workers is None:
        max_workers = min(32, (multiprocessing.cpu_count() or 1) + 4)
    print(f"‚ö° Using parallel processing with {max_workers} workers")

    deleted_count = 0
    converted_count = 0

    # 2. Process in parallel
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = [executor.submit(process_single_file, file_info) for file_info in file_list]
        
        for future in as_completed(futures):
            result = future.result()
            if result['warning']:
                print(result['warning'])
            deleted_count += result['deleted']
            converted_count += result['converted']

    print(f"‚úÖ Data cleaning complete!")
    print(f"   - Deleted/Corrupted: {deleted_count}")
    print(f"   - Converted to RGB: {converted_count}")

# Run cleaning on both datasets
clean_and_verify_data(CONFIG['train_data_dir'])
clean_and_verify_data(CONFIG['test_data_dir'])
```

```python
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15), # Á®çÂæÆÂ¢ûÂä†ÈöæÂ∫¶ÔºåÈò≤Ê≠¢ËøáÊãüÂêà
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

# ÂÅáËÆæ‰Ω†ÁöÑ dataset_assigned Êñá‰ª∂Â§π‰∏ãÊúâ 'train' Âíå 'val' Â≠êÊñá‰ª∂Â§π
try:
    image_datasets = {
        'train': datasets.ImageFolder(CONFIG['train_data_dir'], data_transforms['train']),
        'val': datasets.ImageFolder(CONFIG['test_data_dir'], data_transforms['val'])  # Test data
    }
    
    
    # Mac Ê≥®ÊÑè: num_workers ÂèØ‰ª•ËÆæÁΩÆ‰∏∫ 2-4 ‰ª•Âä†ÈÄüÊï∞ÊçÆÂä†ËΩΩ
    # Â¶ÇÊûúÈÅáÂà∞Â§öËøõÁ®ãÊä•ÈîôÔºåÊîπÂõû num_workers=0
    num_workers = 4  # ÂèØÊ†πÊçÆ CPU Ê†∏ÂøÉÊï∞Ë∞ÉÊï¥ (M4 ÊúâÂ§öÊ†∏)

    # Only enable pinned memory for CUDA (not supported / no-op on MPS/CPU)
    use_pin_memory = True if CONFIG.get('device', torch.device('cpu')).type == 'cuda' else False
    if not use_pin_memory:
        print("Note: pin_memory disabled for non-CUDA device (MPS/CPU).")

    dataloaders = {
        'train': DataLoader(
            image_datasets['train'],
            batch_size=CONFIG['batch_size'],
            shuffle=True,
            num_workers=num_workers,
            pin_memory=use_pin_memory,
            persistent_workers=(num_workers > 0)
        ),
        'val': DataLoader(
            image_datasets['val'],
            batch_size=CONFIG['batch_size'],
            shuffle=False,  # No shuffle for consistent filename order in evaluation
            num_workers=num_workers,
            pin_memory=use_pin_memory,
            persistent_workers=(num_workers > 0)
        )
    }

    dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'val']}
    class_names = image_datasets['train'].classes
    print(f"Classes found: {class_names}")

    # === ÂÆâÂÖ®Ê£ÄÊü• ===
    if len(class_names) != CONFIG['num_classes']:
        print(f"‚ö†Ô∏è WARNING: You set num_classes={CONFIG['num_classes']}, but found {len(class_names)} classes in the folder!")
        print("Please check your dataset folder or update CONFIG['num_classes'].")
    else:
        print(f"‚úÖ Class count matches: {len(class_names)}")
        

except Exception as e:    
    print("ËØ∑Á°Æ‰øù 'dataset_assigned' Êñá‰ª∂Â§πÈáåÊúâ 'train' Âíå 'val' Êñá‰ª∂Â§πÔºåÂπ∂‰∏îÈáåÈù¢ÊúâÂõæÁâáÂàÜÁ±ªÊñá‰ª∂Â§π„ÄÇ")
    print(f"‚ùå Error loading data: {e}")
```

## 3. Model Definition (ResNet18 with Transfer Learning)

```python
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# ÂÜªÁªìÁâπÂæÅÊèêÂèñÂ±Ç
for param in model.parameters():
    param.requires_grad = False

# ‰øÆÊîπÂÖ®ËøûÊé•Â±Ç
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, CONFIG['num_classes'])

# Â∞ÜÊ®°ÂûãÁßªÂä®Âà∞ M4 GPU
model = model.to(CONFIG['device'])

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=CONFIG['learning_rate'])
```

## 4. Training Function

```python
def train_model(model, criterion, optimizer, num_epochs=5, patience=3):
    since = time.time()
    history = {'train_acc': [], 'val_acc': [], 'train_loss': [], 'val_loss': []}
    
    # Early Stopping ÂèòÈáè
    best_val_acc = 0.0
    best_model_wts = None
    patience_counter = 0

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                # ÂÖ≥ÈîÆÔºöÊï∞ÊçÆ‰πüË¶ÅÁßªÂä®Âà∞ MPS
                inputs = inputs.to(CONFIG['device'])
                labels = labels.to(CONFIG['device'])

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            # Use .float() instead of .double() to avoid MPS float64 incompatibility
            epoch_acc = running_corrects.float() / dataset_sizes[phase]

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
            
            if phase == 'train':
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc.item())
            else:
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc.item())
                
                # Early Stopping ÈÄªËæë
                if epoch_acc > best_val_acc:
                    best_val_acc = epoch_acc
                    best_model_wts = model.state_dict().copy()
                    patience_counter = 0
                    print(f'‚úÖ New best model! Val Acc: {best_val_acc:.4f}')
                else:
                    patience_counter += 1
                    print(f'‚è≥ No improvement for {patience_counter}/{patience} epochs')
                    
                    if patience_counter >= patience:
                        print(f'\nüõë Early Stopping triggered! Best Val Acc: {best_val_acc:.4f}')
                        print(f'   Stopped at Epoch {epoch+1}/{num_epochs}')
                        model.load_state_dict(best_model_wts)
                        time_elapsed = time.time() - since
                        print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
                        return model, history

    # Â¶ÇÊûúÊ≤°Êúâ Early StopÔºå‰πüÂä†ËΩΩÊúÄ‰Ω≥Ê®°Âûã
    if best_model_wts is not None:
        model.load_state_dict(best_model_wts)
        print(f'\n‚úÖ Loaded best model with Val Acc: {best_val_acc:.4f}')
    
    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    return model, history
```

## 5. Execute Training

```python
# Â¶ÇÊûúÊï∞ÊçÆÂä†ËΩΩÊàêÂäüÔºåÂºÄÂßãËÆ≠ÁªÉ
if 'dataloaders' in locals():
    trained_model, history = train_model(
        model, 
        criterion, 
        optimizer, 
        num_epochs=CONFIG['num_epochs'],
        patience=CONFIG['early_stop_patience']
    )
else:
    print("‚ùå Data not loaded. Please check the dataset path.")
```

## 7. Confusion Matrix and Classification Report

```python
if 'trained_model' in locals() and 'dataloaders' in locals():
    all_preds = []
    all_labels = []
    all_filenames = []
    idx = 0
    trained_model.eval()
    with torch.no_grad():
        for inputs, labels in dataloaders['val']:
            inputs = inputs.to(CONFIG['device'])
            outputs = trained_model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy()) # MPS tensor ËΩ¨ numpy ÈúÄË¶ÅÂÖàËΩ¨ cpu
            all_labels.extend(labels.cpu().numpy())
            
            # Collect filenames
            batch_size = inputs.size(0)
            for i in range(batch_size):
                path, _ = image_datasets['val'].samples[idx]
                filename = os.path.basename(path)
                all_filenames.append(filename)
                idx += 1

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title('Confusion Matrix')
    plt.show()
    
    print("\nClassification Report:\n")
    print(classification_report(all_labels, all_preds, target_names=class_names))
else:
    print("‚ùå Model not trained or data not loaded.")

# Save predictions to CSV
if 'trained_model' in locals() and 'dataloaders' in locals():
    import pandas as pd
    df_predictions = pd.DataFrame({
        'true_label': [class_names[label] for label in all_labels],
        'predicted_label': [class_names[pred] for pred in all_preds]
    })
    df_predictions.to_csv('predictions_test.csv', index=False)
    print("‚úÖ Predictions saved to 'predictions_test.csv'")
else:
    print("‚ùå Cannot save predictions: Model not trained or data not loaded.")
```

```python
# ==========================================
# Á¥ßÊÄ•Â§áÁî®ÔºöÁâ©ÁêÜÊâ©ÂÖÖÊï∞ÊçÆÈõÜËÑöÊú¨ (Offline Augmentation) [PARALLEL VERSION]
# Áî®‰∫éÔºöÂΩìËÄÅÂ∏àË¶ÅÊ±Ç"ÁîüÊàêÂπ∂‰øùÂ≠ò"Êõ¥Â§öÂõæÁâáÊó∂
# ==========================================
import torchvision.transforms as T

def augment_single_image(args):
    """Augment a single image (for parallel processing)"""
    src_path, save_path, new_filename = args
    
    augmentor = T.Compose([
        T.RandomRotation(20),
        T.RandomHorizontalFlip(p=1.0),
        T.ColorJitter(brightness=0.2, contrast=0.2),
    ])
    
    try:
        with Image.open(src_path) as img:
            if img.mode != 'RGB':
                img = img.convert('RGB')
            aug_img = augmentor(img)
            aug_img.save(save_path)
        return f"‚úì {new_filename}"
    except Exception as e:
        return f"‚úó {new_filename}: {e}"

def augment_and_save(root_dir, target_count=50, max_workers=None):
    """Parallel version of offline augmentation"""
    print(f"üöÄ Starting Parallel Offline Augmentation to reach {target_count} images per class...")
    
    if max_workers is None:
        max_workers = multiprocessing.cpu_count() or 4
    
    tasks = []
    
    for subdir, dirs, files in os.walk(root_dir):
        # Filter for valid images only
        valid_images = [f for f in files if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        current_count = len(valid_images)
        
        if current_count > 0 and current_count < target_count:
            print(f"Processing folder: {subdir} (Current: {current_count})")
            needed = target_count - current_count
            
            for i in range(needed):
                src_file = valid_images[i % current_count]
                src_path = os.path.join(subdir, src_file)
                new_filename = f"{os.path.splitext(src_file)[0]}_aug_{i}.jpg"
                save_path = os.path.join(subdir, new_filename)
                tasks.append((src_path, save_path, new_filename))
    
    print(f"üìä Total augmentation tasks: {len(tasks)}")
    
    if not tasks:
        print("‚úÖ No augmentation needed.")
        return

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = [executor.submit(augment_single_image, task) for task in tasks]
        
        completed = 0
        for future in as_completed(futures):
            _ = future.result() # We just consume the result
            completed += 1
            if completed % 100 == 0:
                print(f"   Progress: {completed}/{len(tasks)} images processed")

    print("‚úÖ Offline Augmentation Complete!")

# Usage:
augment_and_save(os.path.join(CONFIG['data_dir'], 'train'), target_count=50)
```
